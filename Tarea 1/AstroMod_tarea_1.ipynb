# 1. Una partícula de la masa en reposo $m_0$, energía $E_0$ y momento $p_0$ sufre una colisión elástica frontal (es decir, masas de partículas sin alterar) con energía $E$ y momento $P$, dejando la primera partícula con energía $E_1$ y $p_1$. Calcule $P$ y $p_1$, ¿a qué se reducen estas expresiones en el límite clásico? Realice una simulación del fenómeno acá expuesto.

AstroMdo_Tarea#1_parte1.png


Partimos de la conservación del momento de las masas así:
Usando la conservación del momento tenemos que:

\begin{equation}
p_0 = p_1 + P
\end{equation}

En donde ya conocemos el valor de $p_0$, luego usamos la converación de la energía en donde la suma de las energias iniciales de las dos particulas deben ser iguales a la de las energías finales de las dos particulas:

\begin{equation}
E_{0}  + E_{i} = E_{1} + E
\end{equation}

Donde las energías tienen una relación con la masa y su momentum siendo así que:

\begin{equation}
E_{0} = \sqrt{(p_0c)^2 + (m_0c^2)^2}
\\
E_{1} = \sqrt{(p_1c)^2 + (m_0c^2)^2}
\\
E = \sqrt{(Pc)^2 + (Mc^2)^2}
\end{equation}

>$c$ = velocidad de la luz

Y la energía para la masa en reposo es:
\begin{equation}
E_{i}  = Mc^2
\end{equation}

Ahora reemplazamos esas energías y obtenemos que:

\begin{equation}
\sqrt{(p_0c)^2 + (m_0c^2)^2} + Mc^2 = \sqrt{(p_1c)^2 + (m_0c^2)^2} + \sqrt{(Pc)^2 + (Mc^2)^2}
\end{equation}

Luego usando la conservasión del momento y despejamos tenemos que $P = p_0 - p_1$ y resolvemos para $p_1$

\begin{equation}
\sqrt{(p_0c)^2 + (m_0c^2)^2} + Mc^2 = \sqrt{(p_1c)^2 + (m_0c^2)^2} + \sqrt{((p_0-p_1)c)^2 + (Mc^2)^2}
\end{equation}

Luego de hacer su respectiva algebra obtenemos que:

\begin{equation}
p_1 = \frac{m_0^4c^2p_0 + 2M^3p_0\sqrt{c^2(m_0^2c^2+p_0^2)}-2m_0^2Mp_0\sqrt{c^2(m_0^2c^2+p_0^2)}-M^4c^2p_0}{m_0^4c^2-2m_0M^2c^2+M^4c^2-4M^2p_0^2}
\end{equation}

Eso quiere decir que:

\begin{equation}
P = p_0 - \frac{m_0^4c^2p_0 + 2M^3p_0\sqrt{c^2(m_0^2c^2+p_0^2)}-2m_0^2Mp_0\sqrt{c^2(m_0^2c^2+p_0^2)}-M^4c^2p_0}{m_0^4c^2-2m_0M^2c^2+M^4c^2-4M^2p_0^2}
\end{equation}

Comprobemos este resultado.

```python
import numpy as np

#Valores iniciales

m_o = 1.5 #kg
p_o = 2 #kg*m/s
M = 3 #kg
c = 3e+8 #m/s

p_1 = (m_o**2*c**2*p_o+2*M**3*p_o*np.sqrt(c**2*(m_o**2*c**2+p_o**2))-2*m_o**2*M*p_o*np.sqrt(
c**2*(m_o**2*c**2+p_o**2))-M**4*c**2*p_o)/(m_o**4*c**2-2*m_o**2*M**2*c**2+M**4*c**2-4*M**2*p_o**2)

P = p_o-p_1

#definimos una función para las energías
def E(p,m):
    return np.sqrt((p*c)**2+(m*c**2)**2)

e1 = E(p_o,m_o) + M*c**2 #energia incial
e2 = E(p_1,m_o) + E(P,M) #energia final

print(f"Resultado del momento de p1: {p_1} kg*m/s")
print(f"Resultado del momento de P: {P} kg*m/s")
print(f"Resultado de las sumas de energías inciales: {e1} J")
print(f"Resultado de las sumas de energías finales: {e2} J")
```

Observemos los resultados y se nota que tiene sentido físico, comenzando con el signo de $p_1$ vemos que es negativo significando que la dirección del momento ahora va en sentido contrario despues del choque (usar dibujo del principio como guía) mientras que $P$ tiene una dirección positiva y es más grande en magnitud porque hay una transferencia de energia. Luego vemos que las energías sumas de la energías iniciales y finales son iguales. Ahora que comprobamos que el resultado es correcto se procede hacer una simulación del fenomeno

# 2. Muestre que la colisión inelástica, la masa en reposo del objeto combinado es mayor que la suma de las masas en reposo originales. ¿Por cuánto más es mayor la masa? ¿de dónde deriva este aumento? ¿qué significa esto?

AstroMod_tarea#1_parte2.png

Supongamos dos particulas con la misma masa y velocidad. Usando la conservación del momento:

\begin{equation}
P_o = P_f \rightarrow mv\gamma - mv\gamma = 0
\end{equation}

Entonces por lo anterior  $P_f = 0$ luego:

\begin{equation}
P_f = 0 \rightarrow Mv_m\gamma_m = 0
\end{equation}

si $P_f = 0$ eso quiere decir que $v_m = 0$ ó $\gamma_m = 0$ porque $M \neq 0$ luego recordemos que la masa final está en reposo y eso quiere decir que $v_m = 0$ y $\gamma_m = 1$

>Recordemos que $\gamma = \frac{1}{\sqrt{1-\frac{v^2}{c^2}}}$ y $c$ es la velocidad de la luz

Ahora usando la conservación de la energía donde:

\begin{equation}
E_o = E_f \rightarrow m\gamma c^2 + m\gamma c^2 = M\gamma_m c^2 \rightarrow 2m\gamma = M\gamma_m \rightarrow M = 2m\gamma
\end{equation}

Ahora expandamos el término $\gamma$ usando la serie de taylor:

\begin{equation}
\Sigma_{n=0}^\infty \frac{1}{n!} \frac{d^nf(x)}{dx^n} (x-x_o)^n
\end{equation}

Expandimos el $\gamma$ y obtenemos:

\begin{equation}
1+\frac{v^2}{2c^2}+\frac{3v^4}{8c^2}+\frac{5v^6}{16c^6}+.....
\end{equation}

Usamos los dos primero terminos y luego reemplazamos para obtener:

\begin{equation}
M = 2m(1+\frac{v^2}{2c^2}) \rightarrow M = 2(m+ \frac{mv^2}{2c^2}) \rightarrow M = 2(m+ \frac{E_k}{c^2})
\end{equation}

>Energia cinetica $E_k$: $\frac{1}{2}mv^2$

Entonces podemos concluir que la masa resultante es mayor por 2 veces la masas de las dos particulas más dos veces la energía cinetica dividida por la velocidad de la luz. Esto significa que no solo las masas de las partículas aportan a la masa resultante si no que las energías de las mismas tambien corroborando así que la energía es igual a la masa por la velocidad de la luz (obviamente esto es cierto en el caso relativistico, si lo pasamos en casos galileanos también va a funcionar pero la energía no hará un aporte muy grande) $E = Mc^2$. # 3. Dos partículas cuyas masa en reposo son $m_1$ y $m_2$ se mueven a lo largo de una linea recta con velocidades $u_1$ y $u_2$, medidas en la misma dirección. Chocan inelásticamente para formar una nueva partícula. Encuentre la masa en reposo y la velocidad de la nueva partícula. Realice una simulación del fenómeno acá expuesto

AstroMod_tarea#1_parte3.png

Para que dos particulas viajando en la misma dirección choquen la velocidad de una de las particulas tiene que ser mayor que la de la otra, para nosotros $u_1 > u_2$. Primero veamos la conservación del momento de este sistema:

\begin{equation}
p_o = p_f \rightarrow m_1u_1\gamma_1 + m_2u_2\gamma_2 = Mv\gamma_m
\end{equation}

> $\gamma_1 = \frac{1}{\sqrt{1-\frac{u_1^2}{c^2}}}$, $\gamma_2 = \frac{1}{\sqrt{1-\frac{u_2^2}{c^2}}}$ y $\gamma_m = \frac{1}{\sqrt{1-\frac{v^2}{c^2}}}$

Luego con la conservación de la energia obtenemos que:

\begin{equation}
E_o = E_f \rightarrow m_1c^2\gamma_1 + m_2c^2\gamma_2 = Mc^2\gamma_m \rightarrow m_1\gamma_1 + m_2\gamma_2 = M\gamma_m \rightarrow M = \frac{m_1\gamma_1 + m_2\gamma_2}{\gamma_m}
\end{equation}

Luego reemplazamos $M$ en la primera ecuación y obtenemos que:

\begin{equation}
m_1u_1\gamma_1 + m_2u_2\gamma_2 = \frac{m_1\gamma_1 + m_2\gamma_2}{\gamma_m}v\gamma_m \rightarrow v = \frac{m_1u_1\gamma_1 + m_2u_2\gamma_2}{m_1\gamma_1 +m_2\gamma_2}
\end{equation}

Luego al final la masa resultante queda:

\begin{equation}
M = (m_1\gamma_1 + m_2\gamma_2)(\sqrt{1-\frac{(\frac{m_1u_1\gamma_1 + m_2u_2\gamma_2}{m_1\gamma_1 +m_2\gamma_2})^2}{c^2}})
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt

#Particula 1 y 2

#masa

m1 = 15 #kg
m2 = 20 #kg


#velocidad
c = 3e+8 #m/s

u1 = 0.6*c
u2 = 0.5*c

gamma1 = 1/(np.sqrt(1-u1**2/c**2))
gamma2 = 1/(np.sqrt(1-u2**2/c**2))

#particula resultante

v = (m1*u1*gamma1+m2*u2*gamma2)/(m1*gamma1+m2*gamma2)

gamma = 1/(np.sqrt(1-v**2/c**2))

M = (m1*gamma1+m2*gamma2)/gamma
```

```python
#Parametros de gráfica
Nt = 100
t = np.linspace(0,10,Nt) #s
r0 = c #m
y = np.zeros(Nt)

#Ancho de masa 1, 2 y M
Am_1 = m1*1.2
Am_2 = m2*1.2
AM = M*1.2

#primera particula
r1 = np.zeros(Nt)
for i in range(Nt):
    r1[i] = u1*t[i]

#segunda particula
r2 = np.zeros(Nt)
for i in range(Nt):
    r2[i] = r0+u2*t[i]

#Particula resultante
rm = np.zeros(Nt)
for i in range(Nt):
    rm[i] = r2[-1]+v*t[i]
```

```python
import matplotlib.animation as animation
from matplotlib import rc
rc("animation",html="html5")


fig=plt.figure(figsize=(6,6))

punto,=plt.plot(r1,y,'ko', markersize = Am_1, label = "particula m1")
punto1,=plt.plot(r2,y,'ro', markersize = Am_2, label = "particula m2")
titulo=plt.title(f"t = {t[0]:.2f}")
plt.grid();
plt.legend();


def frame(i):
  punto.set_data(r1[i],y[i])
  punto1.set_data(r2[i],y[i])
  titulo.set_text(f"Sistema antes del choque t = {t[i]:.2f}")
  return punto,punto1,titulo

#frame(60)
plt.close(fig)
anim=animation.FuncAnimation(fig,frame,frames=Nt,interval=100,repeat=False)
anim
```

```python
fig=plt.figure(figsize=(6,6))

punto,=plt.plot(rm,y,'go', markersize = AM, label = "particula M")
titulo=plt.title(f"t = {t[0]:.2f}")
plt.grid();
plt.legend();


def frame(i):
  punto.set_data(rm[i],y[i])
  titulo.set_text(f"Sistema despues del choque t = {t[i]:.2f}")
  return punto,titulo

#frame(60)
plt.close(fig)
anim=animation.FuncAnimation(fig,frame,frames=Nt,interval=100,repeat=False)
anim
```
